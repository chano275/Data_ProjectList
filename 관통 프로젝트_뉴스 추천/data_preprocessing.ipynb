```python
import pandas as pd
original_data = './ssafy_dataset_news_2023.csv'
df = pd.read_csv(original_data, sep='|', skipinitialspace=True)# , index_col=0)
df = df.reset_index()
df
```

```python
df = df.dropna() 
df.info() 
df.shape 
```

```python
df['category'].unique()
```

```python
category_counts = df['category'].value_counts()  # 각 테마별 기사 개수 계산
filtered_categories = category_counts[category_counts >= 10000].index  # 10,000개 이상인 카테고리 필터링

filtered_categories
```

```python
categories = ['사회', '정치', '경제', '국제', '산업', '지방', '스포츠', '연예', '증권','문화']
filtered_df = df[df['category'].isin(categories)]  # 해당 카테고리들만 필터링한 데이터프레임 생성

filtered_df = filtered_df.drop(columns=['index'])

filtered_df
```

```python
filtered_df.to_csv('output.csv', index=False)
```
